```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
noise = np.random.rand(100) * 5
x = np.linspace(0,10,100)

a,b = 2,-5
def f(a,b):
  return a* x -b + noise
y = f(a,b)
```

```python
x.shape , y.shape
```

```python
plt.scatter(x,y)
```

```python
# import sklearn.linear_model
# model = sklearn.linear_model.LinearRegression()
```

```python
from sklearn.linear_model import LinearRegression # LinearReg~ 클래스임
model = LinearRegression()
model.fit(x[:,np.newaxis],y)
#model.fit(x.reshape(-1,1),y)
```

```python
new_x = np.linspace(0,10,100)
new_y = model.predict(new_x[:,np.newaxis])
plt.scatter(x,y)
plt.plot(new_x,new_y)
```

```python
model.coef_ , model.intercept_
```

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])
model.fit(X,y)
model.coef_ , model.intercept_
```

```python
from sklearn.preprocessing import PolynomialFeatures
x = np.array([2,3,4])
poly = PolynomialFeatures(degree=3,include_bias=False)
poly
x
```

```python
trans_x = poly.fit_transform(x[:,np.newaxis])
y = 0.5 + np.dot(trans_x, [1.5, -2., 1.])

model.fit(trans_x,y)
model.coef_ , model.intercept_
```

```python
from sklearn.pipeline import make_pipeline
poly_model = make_pipeline(PolynomialFeatures(7),LinearRegression())
```

```python
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)

poly_model.fit(x[:,np.newaxis],y)
new_y = poly_model.predict(new_x[:,np.newaxis])

plt.scatter(x,y)
plt.plot(new_x,new_y)
```
